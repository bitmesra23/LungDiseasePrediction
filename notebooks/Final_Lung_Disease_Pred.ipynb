```python
# =============================================================
# IMPROVED LUNG X-RAY CLASSIFICATION TRAINING
# Goal: push val accuracy from ~79% toward 90%+
# Key idea: your train/val gap was small (80% vs 79%) -> UNDERFITTING,
# not overfitting. Fix = unfreeze more of the backbone + fine-tune,
# not more regularization.
# =============================================================

import os
import copy
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, WeightedRandomSampler

from torchvision import datasets, transforms, models

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tqdm.auto import tqdm

# -------------------------------------------------------------
# 1. PATHS & DATASETS
# -------------------------------------------------------------
DATA_DIR = "/content/my_extracted_data/chest-xray"

TRAIN_DIR = os.path.join(DATA_DIR, "train")
VAL_DIR = os.path.join(DATA_DIR, "val")
TEST_DIR = os.path.join(DATA_DIR, "test")

IMG_SIZE = 224

# X-rays are near-grayscale -> heavy color jitter doesn't help and can hurt.
# Swapped in affine (small translate/scale) + random erasing (simulates
# occlusion/artifacts) which are more useful for medical imaging.
train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.RandomAffine(degrees=0, translate=(0.05, 0.05), scale=(0.95, 1.05)),
    transforms.ColorJitter(brightness=0.15, contrast=0.15),  # mild only
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.25, scale=(0.02, 0.08)),
])

val_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset = datasets.ImageFolder(VAL_DIR, transform=val_transform)
test_dataset = datasets.ImageFolder(TEST_DIR, transform=val_transform)

print(train_dataset.class_to_idx)

# -------------------------------------------------------------
# 2. HANDLE CLASS IMBALANCE (Emphysema has ~20% fewer samples)
# -------------------------------------------------------------
class_counts = np.bincount([label for _, label in train_dataset.samples])
class_weights_np = class_counts.sum() / (len(class_counts) * class_counts)
print("Class weights:", dict(zip(train_dataset.classes, class_weights_np)))

# Option A: weighted sampler (rebalances batches)
sample_weights = [class_weights_np[label] for _, label in train_dataset.samples]
sampler = WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)

BATCH_SIZE = 32

train_loader = DataLoader(
    train_dataset, batch_size=BATCH_SIZE, sampler=sampler,
    num_workers=2, pin_memory=True
)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

# Option B (used below too): class weights baked into the loss as a backstop
class_weights = torch.tensor(class_weights_np, dtype=torch.float32).to(device)

# -------------------------------------------------------------
# 3. MODEL
# -------------------------------------------------------------
def create_densenet121(num_classes=6, freeze_backbone=True):
    model = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)

    for param in model.features.parameters():
        param.requires_grad = not (not freeze_backbone)  # placeholder, set below
    for param in model.features.parameters():
        param.requires_grad = False if freeze_backbone else True

    in_features = model.classifier.in_features
    model.classifier = nn.Sequential(
        nn.Linear(in_features, 512),
        nn.ReLU(),
        nn.BatchNorm1d(512),
        nn.Dropout(0.4),
        nn.Linear(512, num_classes)
    )
    return model


def unfreeze_last_blocks(model, n_blocks=2):
    """Unfreeze the last n dense blocks + final norm for fine-tuning."""
    # DenseNet121 feature children: conv0, norm0, relu0, pool0,
    # denseblock1, transition1, denseblock2, transition2,
    # denseblock3, transition3, denseblock4, norm5
    children = list(model.features.named_children())
    unfreeze_names = [name for name, _ in children[-(n_blocks * 2 + 1):]]  # blocks+transitions+norm5
    for name, module in children:
        if name in unfreeze_names:
            for p in module.parameters():
                p.requires_grad = True
    return model


model = create_densenet121(num_classes=6, freeze_backbone=True)
model = model.to(device)

criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=0.1)
scaler = torch.amp.GradScaler("cuda")

# -------------------------------------------------------------
# 4. TRAIN / VALIDATE LOOPS
# -------------------------------------------------------------
def train_one_epoch(model, loader, optimizer):
    model.train()
    running_loss, correct, total = 0, 0, 0

    for images, labels in tqdm(loader):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()

        with torch.amp.autocast("cuda"):
            outputs = model(images)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # stabilizes fine-tuning
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    return running_loss / len(loader), 100 * correct / total


def validate(model, loader):
    model.eval()
    running_loss, correct, total = 0, 0, 0

    with torch.no_grad():
        for images, labels in tqdm(loader):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    return running_loss / len(loader), 100 * correct / total


def run_training(model, optimizer, scheduler, epochs, history, best_acc, patience=5, tag="phase"):
    patience_counter = 0
    for epoch in range(epochs):
        print(f"\n[{tag}] Epoch {epoch+1}/{epochs}")
        train_loss, train_acc = train_one_epoch(model, train_loader, optimizer)
        val_loss, val_acc = validate(model, val_loader)

        if scheduler is not None:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(val_loss)
            else:
                scheduler.step()

        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        print(f"Train Acc: {train_acc:.2f}% | Val Acc: {val_acc:.2f}% | Val Loss: {val_loss:.4f}")

        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), "best_densenet121.pth")
            print("Model Saved!")
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping (no improvement for {patience} epochs).")
                break

    return best_acc, history


history = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}
best_acc = 0

# -------------------------------------------------------------
# PHASE 1: train only the classifier head (backbone frozen)
# -------------------------------------------------------------
optimizer_phase1 = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model.parameters()), lr=1e-3, weight_decay=1e-4
)
scheduler_phase1 = torch.optim.lr_scheduler.OneCycleLR(
    optimizer_phase1, max_lr=1e-3, epochs=5, steps_per_epoch=len(train_loader)
)

best_acc, history = run_training(
    model, optimizer_phase1, scheduler_phase1, epochs=5,
    history=history, best_acc=best_acc, patience=5, tag="Phase 1 (head only)"
)

# -------------------------------------------------------------
# PHASE 2: unfreeze last 2 dense blocks, fine-tune whole thing
# with a much smaller LR for the backbone (this is where most
# of your accuracy gain will come from)
# -------------------------------------------------------------
model = unfreeze_last_blocks(model, n_blocks=2)

backbone_params = [p for n, p in model.named_parameters() if p.requires_grad and "classifier" not in n]
head_params = [p for n, p in model.named_parameters() if "classifier" in n]

optimizer_phase2 = torch.optim.AdamW([
    {"params": backbone_params, "lr": 1e-5},
    {"params": head_params, "lr": 1e-4},
], weight_decay=1e-4)

scheduler_phase2 = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer_phase2, T_max=25)

best_acc, history = run_training(
    model, optimizer_phase2, scheduler_phase2, epochs=25,
    history=history, best_acc=best_acc, patience=7, tag="Phase 2 (fine-tune)"
)

print(f"\nBest validation accuracy achieved: {best_acc:.2f}%")

# -------------------------------------------------------------
# 5. PLOT CURVES
# -------------------------------------------------------------
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history["train_acc"], label="train")
plt.plot(history["val_acc"], label="val")
plt.title("Accuracy"); plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history["train_loss"], label="train")
plt.plot(history["val_loss"], label="val")
plt.title("Loss"); plt.legend()
plt.tight_layout()
plt.savefig("training_curves.png")
plt.show()

# -------------------------------------------------------------
# 6. TEST WITH TEST-TIME AUGMENTATION (small extra accuracy boost)
# -------------------------------------------------------------
model.load_state_dict(torch.load("best_densenet121.pth"))
model.eval()

def predict_with_tta(model, loader, n_aug=3):
    all_preds, all_labels = [], []
    with torch.no_grad():
        for images, labels in tqdm(loader):
            images = images.to(device)
            probs_sum = torch.zeros(images.size(0), 6).to(device)

            # original
            probs_sum += torch.softmax(model(images), dim=1)
            # horizontal flip
            probs_sum += torch.softmax(model(torch.flip(images, dims=[3])), dim=1)

            preds = torch.argmax(probs_sum, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())
    return np.array(all_preds), np.array(all_labels)

preds, labels = predict_with_tta(model, test_loader)

print("Test Accuracy:", accuracy_score(labels, preds))
print(classification_report(labels, preds, target_names=train_dataset.classes))
print(confusion_matrix(labels, preds))
```
